```python
import matplotlib.pyplot as plt
%matplotlib inline

def generate(n):
    return 0

def load(fname):
    with open(fname, 'r') as f:
        n = int(f.readline())
        segs = []
        for i in range(n):
            segs.append(map(lambda x: int(x), f.readline().split(' ')))
        start = map(lambda x: int(x), f.readline().split(' '))
        finish = map(lambda x: int(x), f.readline().split(' '))
    return segs, start, finish

def visualize(segs, title):
    left, bottom, right, top = segs[0]
    for seg in segs:
        plt.plot([seg[0], seg[2]], [seg[1], seg[3]])
        left = min(left, seg[0], seg[2])
        right = max(right, seg[0], seg[2])
        bottom = min(bottom, seg[1], seg[3])
        top = max(top, seg[1], seg[3])
        
    plt.title(title)
    plt.axis([left-1, right+1, bottom-1, top+1])
    plt.show()    

folder = 'D:\\Temp\\1\\1\\tests\\'
for i in range(1,56):
    fname = ('00%d.in'%i)[-6:]
    segs, start, finish = load(folder + fname)
    visualize(segs, fname)
```
